```python

import numpy as np

PROVIDER_LEVEL_CSV = "../data/outputs/multirun/2026-03-27_14-31-33/1/provider_level.csv"
MISS_THRESHOLD = 0.05
CONST_THRESHOLD = 0.01
CORR_THRESHOLD  = 0.9

META_COLS = {
    "billing_provider_npi", "label", "excldate", "revocation_rsn",
    "cohort_label", "insufficient_history_flag",
    "months_observed", "first_month", "last_month",
    "span_months", "missing_months_within_span",
    "fraction_months_observed", "mean_gap_months", "max_gap_months",
}

pl = pd.read_csv(PROVIDER_LEVEL_CSV, low_memory=False)
feat_cols = [c for c in pl.columns if c not in META_COLS]
X = pl[feat_cols].copy()

# ── 1. Drop high-missingness features ─────────────────────────────────────────
miss_frac = X.isna().mean()
drop_miss = miss_frac[miss_frac > MISS_THRESHOLD].index.tolist()
X = X.drop(columns=drop_miss)
print(f"Dropped {len(drop_miss)} high-missingness features (>{MISS_THRESHOLD:.0%} NaN)  → {X.shape[1]} remain")

# ── 2. Drop near-constant features ────────────────────────────────────────────
std = X.std(skipna=True)
drop_const = std[std < CONST_THRESHOLD].index.tolist()
X = X.drop(columns=drop_const)
print(f"Dropped {len(drop_const)} near-constant features (std < {CONST_THRESHOLD})     → {X.shape[1]} remain")

# ── 3. Correlation prune at 0.9 ───────────────────────────────────────────────
corr = X.corr().abs()
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
drop_corr = [col for col in upper.columns if any(upper[col] > CORR_THRESHOLD)]
X = X.drop(columns=drop_corr)
print(f"Dropped {len(drop_corr)} correlated features (r > {CORR_THRESHOLD})            → {X.shape[1]} remain")

print(f"\nFinal feature set: {X.shape[1]} features across {X.shape[0]} providers")
X.head()
```
